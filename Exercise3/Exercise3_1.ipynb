### Task B

```python
# Parameters:
prob_c = [0.4,0.6]
mean = [-2,4]
variance = [1, 10]

def prob_x_bar_c(x,c):
    #return 1/np.sqrt(2*np.pi*sigma) * np.exp(-(x-mu)**2/(2*sigma))
    return ( 1 / np.sqrt(2* np.pi * variance[c]) ) * np.exp(-0.5 * ((x - mean[c]) ** 2) / variance[c] )

x = np.linspace(-10, 20, 1000)
prob_x = prob_c[0] * prob_x_bar_c(x,0) + prob_c[1] * prob_x_bar_c(x,1)
```

```python
# Plotting results:
plt.plot(x, prob_x)
plt.xlabel('x')
plt.ylabel('p(x)')
```

### Task C

```python
# Creating data from c=1 and c=2:
p_c_1 = np.random.choice(x, 400, p=prob_x_bar_c(x,0)/np.sum(prob_x_bar_c(x,0)))

p_c_2 = np.random.choice(x, 600, p=prob_x_bar_c(x,1)/np.sum(prob_x_bar_c(x,1)))

p_c_total = np.append(p_c_1, p_c_2)


# Plotting results:
plt.scatter(p_c_total, np.zeros_like(p_c_total), marker="*", s = 0.1)
plt.xlabel('x')

np.sum(p_c_total)

```

### Task F

We compute and display the PDF in the plot, then we try with different bin numbers

```python
def norm_hist_pdf(bins,range):
    plt.hist(p_c_total, bins, range, density=True)
    plt.xlabel("x")
    plt.ylabel("p(x|$\Theta$)")
    plt.plot(x, prob_x, c="magenta")
    plt.show()
    
norm_hist_pdf(50,[-10,20])
norm_hist_pdf(100,[-10,20])
norm_hist_pdf(1000,[-10,20])
norm_hist_pdf(10,[-10,20])
```

We can observe that for lower bin numbers we get less accurate PDF that overshoots the values, and the opposite happens for high number of bins histogram.
